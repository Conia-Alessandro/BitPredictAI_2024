```python
df = pd.read_csv('btc-usd-short.csv')
df = df.drop(columns='total_volume')
df = df.drop(columns='market_cap')
df = df.drop(columns='snapped_at')

X1 = df.to_numpy()

X1 = X1.flatten()

bf = pd.DataFrame({'price': X1},  index=pd.date_range('20200728',periods=1347,freq='D'))
extend = 30

bf = pd.DataFrame(
    data=bf,
    index=pd.date_range(
        start='20200728',
        periods=len(df.index) + extend,
        freq='D'
    )
)
```

```python
di = bf.index
bf = bf.reset_index().drop('index', axis=1)
```

```python
bf
```

```python
def func(x, a,b,c,d):
    return a * (x ** 3) + b * (x ** 2) + c * x + d

guess = (0.5, 0.5, 0.5, 0.5)

fit_bf = bf.dropna()

col_params = {}

for col in fit_bf.columns:

    x = fit_bf.index.astype(float).values
    y = fit_bf[col].values

    params = curve_fit(func, x, y, guess)

    col_params[col] = params[0]


for col in bf.columns:

    x = bf[pd.isnull(bf[col])].index.astype(float).values
    
    bf[col][x] = func(x, *col_params[col])

bf.index = di

print(bf)
```
